```python
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import plotly.express as px
import pandas as pd
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()

# Create a DataFrame from the Iris data
iriss = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iriss['target'] = iris.target
class_names = iris.target_names

app = dash.Dash(__name__)

app.layout = html.Div([
    html.H1("Visualisasi Data"),

    html.Label('Variabel X'),
    dcc.Dropdown(
        id='dropdown-x',
        options=[
            {'label': 'Sepal Length', 'value': 'sepal length (cm)'},
            {'label': 'Sepal Width', 'value': 'sepal width (cm)'},
            {'label': 'Petal Length', 'value': 'petal length (cm)'},
            {'label': 'Petal Width', 'value': 'petal width (cm)'}
        ],
        placeholder="Variabel X",
        value='sepal length (cm)'
    ),

    html.Label('Variabel Y'),
    dcc.Dropdown(
        id='dropdown-y',
        options=[
            {'label': 'Sepal Length', 'value': 'sepal length (cm)'},
            {'label': 'Sepal Width', 'value': 'sepal width (cm)'},
            {'label': 'Petal Length', 'value': 'petal length (cm)'},
            {'label': 'Petal Width', 'value': 'petal width (cm)'}
        ],
        value='sepal width (cm)'
    ),

    dcc.Graph(id='scatter-plot')
])

@app.callback(
    Output('scatter-plot', 'figure'),
    [Input('dropdown-x', 'value'),
     Input('dropdown-y', 'value')]
)
def update_graph(x_terpilih, y_terpilih):
    fig = px.scatter(
        iriss,
        x = x_terpilih,
        y = y_terpilih,
        color='target',
        labels={'target': 'Kelas'},
        title=f'Scatter Plot: {x_terpilih} dan {y_terpilih}'
    )
    return fig

if __name__ == '__main__':
    app.run_server(port=2223, debug=True)
```
